```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# variables
pe_den = 1000
att_dim = 256

# constants
seq_len = 500
positions = np.arange(seq_len)
positions = np.reshape(positions, (-1, 1))
even_dims = 2 * np.arange(att_dim / 2)
denominators = 1 / (pe_den ** (even_dims / att_dim))
denominators = np.reshape(denominators, (1, -1))

sin_arguments = np.dot(positions, denominators)

positional_encoding = np.zeros((seq_len, att_dim))
positional_encoding[:, ::2] = np.sin(sin_arguments)
positional_encoding[:, 1::2] = np.cos(sin_arguments)

fig = plt.figure(figsize=(5, 5), dpi=200)
plt.imshow(positional_encoding, aspect='auto')
plt.show()
```
